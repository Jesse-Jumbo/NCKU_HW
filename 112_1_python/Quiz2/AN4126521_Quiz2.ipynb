```python
import random

def calculate_income_tax(net_income: int):
    while True:
      try:
        net_income = int(net_income)
        if net_income < 0:
            raise ValueError("輸入必須大於等於0")
        break  # 如果輸入是有效的數字且大於等於0，則退出循環
      except ValueError as e:
        print("請輸入大於等於0的數字：")
        net_income = input()

    # 0~560,000 ×5％－0
    if net_income <= 560000:
        tax = net_income * 0.05 - 0
    # 560,001~1,260,000×12％－39,200
    elif 560001 <= net_income <= 1260000:
        tax = net_income * 0.12 - 39200
    # 1,260,001~2,520,000×20％－140,000
    elif 1260001 <= net_income <= 2520000:
        tax = net_income * 0.20 - 140000
    # 2,520,001~4,720,000×30％－392,000
    elif 2520001 <= net_income <= 4720000:
        tax = net_income * 0.30 - 392000
    # 4,720,001 以上×40％－864,000
    else:
        tax = net_income * 0.40 - 864000

    return max(tax, 0)

# 測試函數
net_income = random.randrange(0, 5720001+1)
income_tax = calculate_income_tax(net_income)
print(f"當綜合所得稅應納稅額為：{net_income}，綜合所得稅應納稅額：{income_tax} 新臺幣元")
```

# 第二題，有一種數字叫做黑洞數
1. 尋找黑洞數的規則
2. 寫一個程式，驗證所有四位數的情況。
3. 你能夠把你的程式改成驗證5位數的情況嗎？


```python
def find_black_hole(num):
  print(f"輸入 {num}")
  while True:
    num_str = str(num)
    ascending = ''.join(sorted(num_str))
    descending = ''.join(sorted(num_str, reverse=True))
    result = int(descending) - int(ascending)
    if result == num:
      print(f"{num} 是黑洞數！")
      break
    num = result

# 驗證所有四位數的情況
for num in range(1000, 10000):
  find_black_hole(num)
```

```python
def find_black_hole(num, max_iterations=100):
    print(f"輸入 {num}")
    for _ in range(max_iterations):
        num_str = str(num)
        ascending = ''.join(sorted(num_str))
        descending = ''.join(sorted(num_str, reverse=True))
        result = int(descending) - int(ascending)
        if result == num:
            print(f"{num} 是黑洞數！")
            break
        num = result
    else:
        print(f"無法找到 {num} 的黑洞數")

# 驗證所有五位數的情況
for num in range(10000, 100000):
    find_black_hole(num)
```
